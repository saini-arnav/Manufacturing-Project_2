```python
x = result.drop(['pass_fail','Time','start_time','year'], axis=1)
y = result[['pass_fail']]
```

### Feature scaling
### Going with Standard scaler 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
```

```python
x_scaler
# let convert this array to DataFrame
```

```python
x = pd.DataFrame(x_scaler, columns=x.columns[:])
```

```python
x.head()
```

```python
y.value_counts()
# clearly imbalance of data
```

```python
import imblearn
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
x_ros, y_ros = ros.fit_resample(x, y)
```

```python
x_ros.shape
# 595 features , will reduce it
```

```python
# principal component analysis - dimension reduction method
from sklearn.decomposition import PCA
pca = PCA(.95)     # 95% confidence level
x_pca = pca.fit_transform(x_ros)
x_pca.shape
```

```python
# reduced to 145
```

## Spliting the data into training and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_pca, y_ros, test_size=0.3, random_state=1)
```

```python
# !pip install tensorflow
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC # kernel = linear,poly, sigmoid, rbf
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import VotingClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier


import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score 
from sklearn.metrics import f1_score, precision_score, recall_score              
```

```python
rforest = RandomForestClassifier(n_estimators=500, criterion='entropy')
rforest.fit(x_train, y_train)
```

```python
y_pred_train = rforest.predict(x_train)
y_pred_test = rforest.predict(x_test)
```

```python
print(confusion_matrix(y_train, y_pred_train))
print("##########"*10)
print(confusion_matrix(y_test, y_pred_test))
```

```python
print(classification_report(y_train, y_pred_train))
print("##########"*10)
print(classification_report(y_test, y_pred_test))
```

```python
print("Training Accuracy :", accuracy_score(y_train, y_pred_train))
print("##########"*10)
print("Test Accuracy :", accuracy_score(y_test, y_pred_test))
```

# Trying all the models in go 

```python
model1=LogisticRegression()
model2=DecisionTreeClassifier()
model3= SVC() # kernel = linear,poly, sigmoid, rbf
model4 = GradientBoostingClassifier()
model5= GaussianNB()
model6= BernoulliNB()
model7=RandomForestClassifier()
```

```python
l=[model1,model2,model3,model4,model5,model6,model7]


def models(x):
    for i in l:
        i.fit(x_train, y_train)    
        y_pred_train = i.predict(x_train)
        y_pred_test = i.predict(x_test)
        print(f"Result for {i}:")
        print("Training Accuracy :", accuracy_score(y_train, y_pred_train))
        print("Test Accuracy :", accuracy_score(y_test, y_pred_test))
        print('#######'*10)
```

```python
models(l)
```
